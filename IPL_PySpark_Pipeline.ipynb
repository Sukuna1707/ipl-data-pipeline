```python
# Cell 1 - Load data
from pyspark.sql.functions import when, col, round, sum, count

df = spark.read.table("matches")
print(f"Total matches: {df.count()}")
df.show(5)

# Cell 2 - Top 5 winning teams
df.groupBy("winner") \
  .count() \
  .orderBy("count", ascending=False) \
  .show(5)

# Cell 3 - Toss advantage analysis
df.groupBy("toss_decision") \
  .agg(
    count("*").alias("total"),
    sum(when(col("toss_winner") == col("winner"), 1).otherwise(0)).alias("won_after_toss")
  ) \
  .withColumn("win_percentage", round(col("won_after_toss") * 100 / col("total"), 2)) \
  .show()

# Cell 4 - Matches per season
df.groupBy("Season") \
  .count() \
  .orderBy("Season") \
  .show()

# Cell 5 - Hyderabad matches
df.filter(col("city") == "Hyderabad") \
  .groupBy("Season") \
  .count() \
  .orderBy("Season") \
  .show()

# Cell 6 - Player of the match awards
df.groupBy("player_of_match") \
  .count() \
  .orderBy("count", ascending=False) \
  .show(10)
```
